# 기본자료형 연습문제

## 문제 1: 정수형 연습
- 10진수, 2진수, 8진수, 16진수로 숫자 255를 표현하고 출력하세요.
- 각 진법의 타입을 확인하세요.

## 문제 2: 실수형 연습
- 3.14159를 변수에 저장하고, 소수점 둘째 자리까지 반올림하여 출력하세요.
- 지수 표기법으로 1.23e-4를 표현하고 출력하세요.

## 문제 3: 문자열 연습
- "Hello, World!" 문자열을 생성하고 다음을 수행하세요:
  - 문자열의 길이 구하기
  - 대문자로 변환
  - 소문자로 변환
  - "World"를 "Python"으로 바꾸기

## 문제 4: 불린형 연습
- 다음 조건들을 불린형으로 평가하고 결과를 출력하세요:
  - 10 > 5
  - 3 == 3.0
  - "hello" == "Hello"
  - 0 and 1
  - 0 or 1

## 문제 5: 타입 변환 연습
- 문자열 "123"을 정수로 변환
- 정수 456을 문자열로 변환
- 실수 3.14를 정수로 변환
- 정수 100을 실수로 변환
각 변환 결과를 출력하세요.

## 문제 6: 리스트 기본 연습
- 1부터 10까지의 숫자를 담은 리스트를 생성하세요.
- 리스트의 첫 번째와 마지막 요소를 출력하세요.
- 리스트에 11을 추가하고 결과를 출력하세요.

## 문제 7: 딕셔너리 기본 연습
- 이름, 나이, 직업을 키로 하는 딕셔너리를 생성하세요.
- 딕셔너리의 모든 키를 출력하세요.
- 딕셔너리의 모든 값을 출력하세요.
- 새로운 키-값 쌍을 추가하세요.
